```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
try:
  df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2222/Mall_Customers.csv')
except FileNotFoundError:
  print("Error: File not found. Please check the file path.")
  exit()


# Display basic statistics
print("Basic Statistics:")
print(df.describe())


# Histograms for numerical features
print("\nHistograms of Numerical Features:")
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(df['Annual Income (k$)'], bins=20)
plt.title('Annual Income Distribution')

plt.subplot(2, 2, 2)
plt.hist(df['Spending Score (1-100)'], bins=20)
plt.title('Spending Score Distribution')

plt.subplot(2, 2, 3)
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')

plt.tight_layout()
plt.show()

# Pairplot for relationships between numerical features
print("\nPairplot of Numerical Features:")
sns.pairplot(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
plt.show()

# Boxplots for detecting outliers
print("\nBoxplots for Outlier Detection:")
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.boxplot(y=df['Annual Income (k$)'])
plt.title('Annual Income Boxplot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['Spending Score (1-100)'])
plt.title('Spending Score Boxplot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Age'])
plt.title('Age Boxplot')

plt.tight_layout()
plt.show()

# Correlation matrix
print("\nCorrelation Matrix:")
correlation_matrix = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

```
